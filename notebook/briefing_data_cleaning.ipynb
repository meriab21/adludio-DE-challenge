```python
#cleaning missing values
briefing_data = briefing_data.drop('Delivery Requirements (Black/Audience/White List)', axis=1)
```

```python
#cleaning data by filling missing values 
briefing_data['Buy Rate (CPE)']= briefing_data['Buy Rate (CPE)'].fillna('Not Avaliable')
briefing_data['Volume Agreed']= briefing_data['Volume Agreed'].fillna('Not Avaliable')
briefing_data['Gross Cost/Budget']= briefing_data['Gross Cost/Budget'].fillna('Not Avaliable')
briefing_data['Agency Fee']= briefing_data['Agency Fee'].fillna('Not Avaliable')
briefing_data['Percentage']= briefing_data['Percentage'].fillna('Not Avaliable')
briefing_data['Flat Fee'] = briefing_data['Flat Fee'].fillna('Not Available')
briefing_data['Net Cost']= briefing_data['Net Cost'].fillna('Not Avaliable')
```
